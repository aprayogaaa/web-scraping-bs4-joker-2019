Import Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

HTTP Request

```python
web_url = 'https://www.metacritic.com/movie/joker/user-reviews'
headers = {"User-Agent": "Mozilla/5.0 (Linux; U; Android 4.2.2; he-il; NEO-X5-116A Build/JDQ39) AppleWebKit/534.30 ("
                         "KHTML, like Gecko) Version/4.0 Safari/534.30"}
```

Get Request

```python
response = requests.get(web_url, headers=headers) 
```

Status Code

```python
response.status_code
```

Target Necessary Data

- Review Name
- Review Description
- Review Rating
- Review Thumbsup
- Review Date

Fo-Loop to Put Everything Together

Positve review

```python
review_name = []
review_description = []
review_rating = []
review_thumbsup = []
review_date = []

for i in range(0, 28):

    webpage = 'https://www.metacritic.com/movie/joker/user-reviews?dist=positive&page=' + str(i)
    
    headers = {"User-Agent": "Mozilla/5.0 (Linux; U; Android 4.2.2; he-il; NEO-X5-116A Build/JDQ39) AppleWebKit/534.30 ("
                         "KHTML, like Gecko) Version/4.0 Safari/534.30"}
    
    response = requests.get(webpage, headers=headers) 

    soup = BeautifulSoup(response.content, 'html.parser')

    results = soup.find_all('div', {'class':'review pad_top1'})

    for result in results:
        try:
            review_name.append(result.find('span', {'class':'author'}).get_text())
        except:
            review_name.append('n/a')
        try:
            review_description.append(result.find('div', {'class':'review_body'}).get_text().strip())
        except:
            review_description.append('n/a')
        try:
            review_rating.append(result.find('div', {'class':'metascore_w user large movie positive indiv'}).get_text())
        except:
            review_rating.append('10')
        try:
            review_thumbsup.append(result.find('span', {'class':'count'}).get_text())
        except:
            review_thumbsup.append('n/a')
        try:
            review_date.append(result.find('span', {'class':'date'}).get_text())
        except:
            review_date.append('n/a')
        
```

```python
movie_review_1 = pd.DataFrame({'Username': review_name, 
                             'Review': review_description, 
                             'Rating':review_rating,
                             'Thumbsup': review_thumbsup,
                             'Date': review_date})
```
